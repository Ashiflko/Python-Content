```python
from  bs4 import BeautifulSoup
import requests
```

```python
url="https://www.imdb.com/chart/top/?ref_=nv_mv_250"
```

```python
data= requests.get(url)
```

```python
Soup=BeautifulSoup(data.text)
```

```python
Soup.find('h1')
```

```python
Soup.find('h1').text
```

```python
rows=Soup.find('tbody').findAll('tr')
```

```python
len(rows)
```

```python
for movie in rows :
    print(movie.find('td',{'class':'titleColumn'}).find('a').text)
    print(movie.find('td',{'class':'ratingColumn'}).text)
    print(movie.find('span',{'class':'secondaryInfo'}).text)
```

```python
movie_details=[]
for movie in rows:
    detail={}
    detail['name']=movie.find('td',{'class':'titleColumn'}).find('a').text
    detail['year']=movie.find('td',{'class':'titleColumn'}).find('span').text.strip('()')
    detail['Rating']=movie.find('td',{'class':'imdbRating'}).find('strong').text
    movie_details.append(detail)
```

```python
# 
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(movie_details)
```
